```python
import numpy as np
import pandas as pd
import urllib.parse
import requests
```

```python
repo_owner = "post-no-preference"
repo_name = "earnings"
query = "SHOW TABLES"

# The most reliable URL format for DoltHub's SQL API
url = f"https://www.dolthub.com/api/v1/sql/{repo_owner}/{repo_name}/master"

params = {'q': query}

try:
    response = requests.get(url, params=params)
    response.raise_for_status() # This will catch the 404 immediately
    
    data = response.json()
    
    # DoltHub returns a 'rows' list inside the JSON
    if 'rows' in data:
        df = pd.DataFrame(data['rows'])
        print("Available Tables:")
        print(df)
    else:
        print("Query successful but no rows returned.")
        print(data)

except requests.exceptions.HTTPError as err:
    print(f"HTTP Error: {err}")
except Exception as e:
    print(f"An error occurred: {e}")
```
